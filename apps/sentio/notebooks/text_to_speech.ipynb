## Bloom voice — Qwen3 voice clone (Clara)

Uses **Qwen3 Base** with `ref_audio` + verbatim `ref_text`. CustomVoice presets are still available via `set_voice()`.

Post-processing: high-shelf EQ → compression → limiter.

```python
import sys
from pathlib import Path

sys.path.insert(0, str(Path("..").resolve()))

from qwen3_tts import DEFAULT_CLARA_REF_TEXT, Qwen3TTS

tts = Qwen3TTS()

tts.set_voice_clone(
    ref_audio=Path("voices/clara.wav"),
    ref_text=DEFAULT_CLARA_REF_TEXT,
)
```

```python
import time

import numpy as np
from IPython.display import Audio

from voice_post import SAMPLE_RATE

text = (
    "Hey. I'm Bloom — your private AI assistant. "
    "Everything stays on your network, encrypted, off the grid. "
    "No cloud uploads, no data harvesting, no one listening in. "
    "Just you, your devices, and me running local in the background. "
    "Lights, locks, schedules, questions — whatever you need, I handle it quietly. "
    "Your home. Your rules. What's the move?"
)

t0 = time.perf_counter()
audio = tts.synthesize(
    text,
    temperature=0.9,
    high_boost_db=5.0,
    air_boost_db=2.5,
    compress_threshold_db=-20,
    compress_ratio=2.5,
    compress_makeup_db=2.0,
)
dt = time.perf_counter() - t0

print(f"Generated {len(audio) / SAMPLE_RATE:.2f}s audio in {dt:.1f}s")
Audio(data=audio.astype(np.float32), rate=SAMPLE_RATE)
```
